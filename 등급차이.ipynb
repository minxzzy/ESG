## t+1 등급차이

```python
e1_grade_sub = pd.read_csv('csv/재무비율/최종/등급차이/환경_1기_등급차이포함.csv', index_col=0).reset_index(drop=True)
s1_grade_sub = pd.read_csv('csv/재무비율/최종/등급차이/사회_1기_등급차이포함.csv', index_col=0).reset_index(drop=True)
g1_grade_sub = pd.read_csv('csv/재무비율/최종/등급차이/지배구조_1기_등급차이포함.csv', index_col=0).reset_index(drop=True)
t1_grade_sub = pd.read_csv('csv/재무비율/최종/등급차이/통합등급_1기_등급차이포함.csv', index_col=0).reset_index(drop=True)
esg2_grade_sub = pd.read_csv('csv/재무비율/최종/등급차이/esg2기_등급차이포함.csv', index_col=0).reset_index(drop=True)
```

```python
e1_grade_sub
```

```python
e1_grade_sub['t+1년도_환경_등급차이'] = e1_grade_sub.groupby('종목명')['환경_등급차이'].shift(1)
e1_grade_sub.dropna(inplace=True)
e1_grade_sub.reset_index(drop=True, inplace=True)

s1_grade_sub['t+1년도_환경_등급차이'] = s1_grade_sub.groupby('종목명')['사회_등급차이'].shift(1)
s1_grade_sub.dropna(inplace=True)
s1_grade_sub.reset_index(drop=True, inplace=True)

g1_grade_sub['t+1년도_환경_등급차이'] = g1_grade_sub.groupby('종목명')['지배구조_등급차이'].shift(1)
g1_grade_sub.dropna(inplace=True)
g1_grade_sub.reset_index(drop=True, inplace=True)

t1_grade_sub['t+1년도_환경_등급차이'] = t1_grade_sub.groupby('종목명')['통합_등급차이'].shift(1)
t1_grade_sub.dropna(inplace=True)
t1_grade_sub.reset_index(drop=True, inplace=True)

esg2_grade_sub['t+1년도_환경_등급차이'] = esg2_grade_sub.groupby('종목명')['환경_등급차이'].shift(1)
esg2_grade_sub['t+1년도_사회_등급차이'] = esg2_grade_sub.groupby('종목명')['사회_등급차이'].shift(1)
esg2_grade_sub['t+1년도_지배구조_등급차이'] = esg2_grade_sub.groupby('종목명')['지배구조_등급차이'].shift(1)
esg2_grade_sub['t+1년도_통합_등급차이'] = esg2_grade_sub.groupby('종목명')['통합_등급차이'].shift(1)
esg2_grade_sub.dropna(inplace=True)
esg2_grade_sub.reset_index(drop=True, inplace=True)
esg2_grade_sub
```
