## t+1 년도 esg 등급 데이터프레임

```python
e1_shift = e1_total.copy()
e1_shift['t+1년도_환경등급'] = e1_total.groupby(['종목명','업종명','종목코드'])['환경_등급'].shift(-1)
e1_shift.dropna(inplace=True)
e1_shift.reset_index(drop=True, inplace=True)

s1_shift = s1_total.copy()
s1_shift['t+1년도_사회등급'] = s1_total.groupby(['종목명','업종명','종목코드'])['사회_등급'].shift(-1)
s1_shift.dropna(inplace=True)
s1_shift.reset_index(drop=True, inplace=True)

g1_shift = g1_total.copy()
g1_shift['t+1년도_지배구조등급'] = g1_total.groupby(['종목명','업종명','종목코드'])['지배구조_등급'].shift(-1)
g1_shift.dropna(inplace=True)
g1_shift.reset_index(drop=True, inplace=True)

t1_shift = t1_total.copy()
t1_shift['t+1년도_통합등급'] = t1_total.groupby(['종목명','업종명','종목코드'])['통합_등급'].shift(-1)
t1_shift.dropna(inplace=True)
t1_shift.reset_index(drop=True, inplace=True)

esg2_shift = esg2_total.copy()
esg2_shift[['t+1년도_환경등급','t+1년도_사회등급','t+1년도_지배구조등급','t+1년도_통합등급']] = esg2_total.groupby(['종목명','업종명','종목코드'])[['환경_등급','사회_등급','지배구조_등급','통합_등급']].shift(-1)
esg2_shift.dropna(inplace=True)
esg2_shift.reset_index(drop=True, inplace=True)
```
