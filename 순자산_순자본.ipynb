```python
import pandas as pd

df = pd.read_csv('csv/총자본_총부채_총자산.csv')
df
```

```python
df.isna().sum()
```

```python
df = df.dropna().reset_index(drop=True)
df
```

```python
df.isna().sum()
```

```python
df['총자산(천원)']
```

```python
df['총자산(천원)'] = df['총자산(천원)'].apply(lambda x : x.replace(',', ''))
df['총자산(천원)']
```

```python
df['총부채(천원)'] = df['총부채(천원)'].apply(lambda x : x.replace(',', ''))
df['총부채(천원)']
```

```python
df['순자산'] = df['총자산(천원)'].apply(lambda x : int(x)) - df['총부채(천원)'].apply(lambda x : int(x))
df['순자산']
```
