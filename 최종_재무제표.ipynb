```python
import pandas as pd
```

### 리스트 작업

```python
list_1 = pd.read_csv('csv/종목리스트/최종_1기_리스트_금융업제외.csv', index_col=0)
list_2 = pd.read_csv('csv/종목리스트/최종_2기_리스트_금융업제외.csv', index_col=0)
```

```python
full = pd.read_csv('csv/재무비율/최최최최종_제무제표.csv')
full
```

```python
full_1 = full.loc[(full['회계연도'] >= 2010) & (full['회계연도'] <= 2018)]
full_2 = full.loc[(full['회계연도'] >= 2019) & (full['회계연도'] <= 2023)]

full_1 = pd.merge(list_1, full_1, on='종목명')
full_2 = pd.merge(list_2, full_2, on='종목명')

full_1.reset_index(drop=True, inplace=True)
full_2.reset_index(drop=True, inplace=True)
```

```python
full_1.to_csv('csv/재무비율/최종_재무제표_1기.csv', encoding='utf-8-sig')
full_2.to_csv('csv/재무비율/최종_재무제표_2기.csv', encoding='utf-8-sig')
```

### 한글값 처리 후 (ex)완전잠식, 적전 등)

```python
full_1 = pd.read_csv('csv/재무비율/최최최최종_재무제표_1기.csv', index_col=0)
full_2 = pd.read_csv('csv/재무비율/최최최최종_재무제표_2기.csv', index_col=0)
```

```python
full_1 = full_1.sort_values(by=['종목명','회계연도']).reset_index(drop=True)
full_2 = full_2.sort_values(by=['종목명','회계연도']).reset_index(drop=True)
```

```python
full_1.columns
```

```python
list = ['총자산(천원)', '유동자산(천원)', '매출채권(천원)', '재고자산(천원)', '유형자산(천원)',
        '부채및자본총계(천원)', '총부채(천원)', '유동부채(천원)', '매입채무(천원)', '총자본(천원)',
        '매출액(천원)', '영업이익(천원)', '당기순이익(천원)',
        '영업활동으로인한현금흐름(천원)', '투자활동으로인한현금흐름(천원)', '재무활동으로인한현금흐름(천원)']

for i in list : 
    full_1[i].apply(lambda x : int(x))
    full_2[i].apply(lambda x : int(x))
```

```python
list = ['총자산(천원)', '유동자산(천원)', '매출채권(천원)', '재고자산(천원)', '유형자산(천원)',
        '부채및자본총계(천원)', '총부채(천원)', '유동부채(천원)', '매입채무(천원)', '총자본(천원)',
        '매출액(천원)', '영업이익(천원)', '당기순이익(천원)',
        '영업활동으로인한현금흐름(천원)', '투자활동으로인한현금흐름(천원)', '재무활동으로인한현금흐름(천원)']

for i in list :
    full_1[f'{i}_증가율(%)'] = full_1.groupby('종목명')[i].pct_change() * 100
    full_2[f'{i}_증가율(%)'] = full_2.groupby('종목명')[i].pct_change() * 100
full_2
```

```python
rotateList = ['총자산(천원)','총자본(천원)','총부채(천원)']

for i in rotateList :
    full_1[f'{i}_회전율'] = full_1['매출액(천원)'].astype(float) / full_1[i] 
    full_2[f'{i}_회전율'] = full_2['매출액(천원)'].astype(float) / full_2[i] 
full_2['총자산(천원)_회전율']
```
