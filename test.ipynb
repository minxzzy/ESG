```python
import pandas as pd

esg_2020 = pd.read_csv('2020_esg.csv', index_col=0)
esg_2021 = pd.read_csv('2021_esg.csv', index_col=0)
esg_2022 = pd.read_csv('2022_esg.csv', index_col=0)
esg_2023 = pd.read_csv('2023_esg.csv', index_col=0)
esg_2024 = pd.read_csv('2024_esg.csv', index_col=0)
```

```python
def word(df, year) :
    df['연도'] = year
    df['회사명'] = df['회사명'].str.strip(f'{year} N N')
    df['종합등급'] = df['종합등급'].str.strip('KCGS')
    df['환경'] = df['환경'].str.strip('환경')
    df['사회'] = df['사회'].str.strip('사회')
    df['지배구조'] = df['지배구조'].str.strip('지배구조')

    return df
```

```python
esg_2020 = word(esg_2020, 2020)
esg_2021 = word(esg_2021, 2021)
esg_2022 = word(esg_2022, 2022)
esg_2023 = word(esg_2023, 2023)
esg_2024 = word(esg_2024, 2024)
```
