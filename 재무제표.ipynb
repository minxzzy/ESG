```python
import pandas as pd
```

```python
list_1 = pd.read_csv('csv/종목리스트/최종_1기_리스트_금융업제외.csv', index_col=0)
list_2 = pd.read_csv('csv/종목리스트/최종_2기_리스트_금융업제외.csv', index_col=0)
```

```python
full = pd.read_csv('csv/재무제표.csv')
full
```

```python
full_1 = full.loc[(full['회계년'] >= 2010) & (full['회계년'] <= 2018)]
full_2 = full.loc[(full['회계년'] >= 2019) & (full['회계년'] <= 2023)]

full_1 = pd.merge(list_1, full_1, on='종목명')
full_2 = pd.merge(list_2, full_2, on='종목명')

full_1.reset_index(drop=True, inplace=True)
full_2.reset_index(drop=True, inplace=True)
```

```python
full.columns
```

```python
def increase(df, _col) :
    df[_col] = df[_col].str.replace(',','').astype(float)
    df[f'{_col}증가율(%)'] = df.groupby('종목명')[_col].pct_change() * 100
```

```python
def convert_num(df, _col) :
    df[_col] = df[_col].str.replace(',','').astype(float)
```
