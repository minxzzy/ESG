```python
import pandas as pd
import matplotlib.pyplot as plt

# 주어진 데이터
data = {
    '업종명': ['화학', '전기전자', '운수장비', '유통업', '기계', '철강금속', '운수창고업', '기타금융', '건설업', 
            '의약품', '서비스업', '음식료품', '종이목재', '비금속광물', '기타제조업', '섬유의복', 
            '전기가스업', '의료정밀', '통신업'],
    'Count': [301, 182, 161, 119, 105, 98, 84, 84, 84, 77, 70, 56, 56, 42, 28, 21, 21, 14, 14]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
plt.barh(df['업종명'], df['Count'], color='skyblue')

# 그래프 제목과 축 레이블 설정
plt.title('1기 업종 분포')
plt.xlabel('Count')
plt.ylabel('업종명')

# 그래프 출력
plt.show()
```
